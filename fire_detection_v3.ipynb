## Understand Dataset Structure

### Subtask:
Inspect the downloaded dataset to understand its directory structure, especially focusing on where images and labels are stored.


```python
import os

# Path to the 'data' directory
data_path = os.path.join(path, 'data')

# List contents of the 'data' directory
print(f"Contents of '{data_path}':")
print(os.listdir(data_path))
```

**Reasoning**:
Now that we know the main subdirectories ('train', 'val', 'test'), the next step is to list the contents of each of these subdirectories to find 'images' and 'labels' folders, as per the instructions.



```python
import os

# List contents of 'train', 'val', and 'test' directories
for subset in ['train', 'val', 'test']:
    subset_path = os.path.join(data_path, subset)
    print(f"\nContents of '{subset_path}':")
    print(os.listdir(subset_path))
```

**Reasoning**:
To further inspect the dataset structure as per the instructions, I will now list the contents of a sample 'images' folder and a sample 'labels' folder (e.g., 'train/images' and 'train/labels') to confirm their presence and see a few file names.



```python
import os

# Path to the 'train' subset
train_path = os.path.join(data_path, 'train')

# List contents of 'train/images'
train_images_path = os.path.join(train_path, 'images')
print(f"\nContents of '{train_images_path}' (first 5 files):")
print(os.listdir(train_images_path)[:5])

# List contents of 'train/labels'
train_labels_path = os.path.join(train_path, 'labels')
print(f"\nContents of '{train_labels_path}' (first 5 files):")
print(os.listdir(train_labels_path)[:5])
```

## Review data.yaml

### Subtask:
Examine the contents of the `data.yaml` file, which is crucial for YOLOv8 configuration. This includes verifying paths to training/validation/test images and labels, and the class names.


**Reasoning**:
To understand the dataset configuration, I will first import the `yaml` library, construct the full path to the `data.yaml` file, read its content, parse it using `yaml.safe_load()`, and then print the resulting dictionary to review its structure and verify the paths and class information.



```python
import yaml
import os

# Construct the full path to the data.yaml file
data_yaml_path = os.path.join(path, 'data.yaml')

# Open and read the data.yaml file
with open(data_yaml_path, 'r') as file:
    data_yaml = yaml.safe_load(file)

# Print the content of the parsed data.yaml dictionary
print("Content of data.yaml:")
print(data_yaml)
```

## Verify Image and Label Paths

### Subtask:
Check if the image and label paths specified in the `data.yaml` file correctly point to the actual locations of the data within the dataset.


**Reasoning**:
To verify the image and label paths, I will iterate through the 'train', 'val', and 'test' subsets, construct their absolute image and label paths based on the dataset root and `data.yaml` entries, and then check if these paths actually exist using `os.path.exists()`.



```python
import os

# Base path of the dataset
dataset_root = path # From previous cells, this is '/kaggle/input/smoke-fire-detection-yolo'

print(f"Verifying paths relative to dataset root: {dataset_root}\n")

# Iterate through 'train', 'val', 'test' to verify paths
for subset_name in ['train', 'val', 'test']:
    # Get relative image path from data.yaml
    relative_images_path = data_yaml[subset_name]

    # Construct full absolute image path
    full_images_path = os.path.join(dataset_root, relative_images_path)

    # Construct full absolute label path by replacing 'images' with 'labels'
    full_labels_path = full_images_path.replace('images', 'labels')

    # Verify existence and print status
    print(f"{subset_name.capitalize()} images path: {full_images_path} - Exists: {os.path.exists(full_images_path)}")
    print(f"{subset_name.capitalize()} labels path: {full_labels_path} - Exists: {os.path.exists(full_labels_path)}\n")
```
